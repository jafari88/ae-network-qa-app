```python
import os
import json

# Set your directory here
network_dir = r"C:\Users\jafar\grok_updated_network14jul_pe05_001Pdecay_flt_fe20_w14\network_snapshots"

# List all node and edge JSON files
files = os.listdir(network_dir)
node_files = [f for f in files if f.endswith('_nodes.json')]
edge_files = [f for f in files if f.endswith('_edges.json')]

print("Node files found:", node_files)
print("Edge files found:", edge_files)

# Pick the first snapshot as an example
if node_files and edge_files:
    node_path = os.path.join(network_dir, node_files[0])
    edge_path = os.path.join(network_dir, edge_files[0])

    with open(node_path, 'r') as f:
        nodes = json.load(f)
    with open(edge_path, 'r') as f:
        edges = json.load(f)

    print(f"\nFirst {len(nodes)} nodes (showing up to 5):")
    for n in nodes[:5]:
        print(n)

    print(f"\nFirst {len(edges)} edges (showing up to 5):")
    for e in edges[:5]:
        print(e)
    
    # NEW: Find highest weight edge
    print("\n" + "="*50)
    print("HIGHEST WEIGHT EDGE ANALYSIS")
    print("="*50)
    
    # Find the edge with maximum weight
    max_weight_edge = max(edges, key=lambda x: x['weight'])
    max_weight = max_weight_edge['weight']
    
    print(f"üèÜ Highest Weight Edge:")
    print(f"   Weight: {max_weight:,.1f}")
    print(f"   Edge: {max_weight_edge['AE1']} ‚Üî {max_weight_edge['AE2']}")
    
    # Show top 10 highest weight edges
    print(f"\nüìä Top 10 Highest Weight Edges:")
    top_edges = sorted(edges, key=lambda x: x['weight'], reverse=True)[:10]
    
    for i, edge in enumerate(top_edges, 1):
        print(f"   {i:2d}. {edge['AE1']} ‚Üî {edge['AE2']} (weight: {edge['weight']:,.1f})")
    
    # Show some statistics
    weights = [edge['weight'] for edge in edges]
    print(f"\nüìà Edge Weight Statistics:")
    print(f"   Total edges: {len(edges):,}")
    print(f"   Average weight: {sum(weights)/len(weights):,.1f}")
    print(f"   Minimum weight: {min(weights):,.1f}")
    print(f"   Maximum weight: {max(weights):,.1f}")
    
    # Find edges containing a specific AE (e.g., DIARRHOEA)
    print(f"\nÔøΩÔøΩ Edges containing 'DIARRHOEA' (top 10 by weight):")
    diarrhoea_edges = [edge for edge in edges if 'DIARRHOEA' in edge['AE1'] or 'DIARRHOEA' in edge['AE2']]
    if diarrhoea_edges:
        top_diarrhoea_edges = sorted(diarrhoea_edges, key=lambda x: x['weight'], reverse=True)[:10]
        for i, edge in enumerate(top_diarrhoea_edges, 1):
            print(f"   {i:2d}. {edge['AE1']} ‚Üî {edge['AE2']} (weight: {edge['weight']:,.1f})")
    else:
        print("   No edges found containing 'DIARRHOEA'")
    
else:
    print("No node or edge JSON files found in the directory.")
```

```python
import streamlit as st
import pandas as pd
import networkx as nx
import json

st.title("AE Network Q&A App")

nodes_file = st.file_uploader("Upload nodes JSON", type="json")
edges_file = st.file_uploader("Upload edges JSON", type="json")

G = None
if nodes_file and edges_file:
    nodes = pd.read_json(nodes_file)
    edges = pd.read_json(edges_file)
    G = nx.Graph()
    for _, row in nodes.iterrows():
        G.add_node(row['AE'], **row.to_dict())
    for _, row in edges.iterrows():
        G.add_edge(row['AE1'], row['AE2'], **row.to_dict())
    st.success(f"Loaded {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")

    question = st.text_input("Ask a question about the network (e.g., 'highest weight', 'top 5 degree', 'list communities'):")

    if question:
        question = question.lower()
        answer = None

        # Highest weight edge
        if "highest weight" in question or "max weight" in question:
            if 'weight' in edges.columns:
                max_row = edges.loc[edges['weight'].idxmax()]
                answer = f"Highest weight edge: {max_row['AE1']} - {max_row['AE2']} (weight={max_row['weight']})"
            else:
                answer = "No 'weight' column found in edges."

        # Top N nodes by degree
        elif "top" in question and "degree" in question:
            import re
            m = re.search(r"top (\d+)", question)
            n = int(m.group(1)) if m else 5
            degree_df = pd.DataFrame(G.degree, columns=['AE', 'degree']).sort_values('degree', ascending=False).head(n)
            answer = degree_df

        # Top N nodes by centrality
        elif "top" in question and "centrality" in question:
            import re
            m = re.search(r"top (\d+)", question)
            n = int(m.group(1)) if m else 5
            dc = nx.degree_centrality(G)
            dc_df = pd.DataFrame(list(dc.items()), columns=['AE', 'degree_centrality']).sort_values('degree_centrality', ascending=False).head(n)
            answer = dc_df

        # List all communities (if present)
        elif "community" in question:
            comms = nodes['community'].unique() if 'community' in nodes.columns else None
            if comms is not None:
                answer = f"Communities detected: {list(comms)}"
            else:
                answer = "No community information found in nodes."

        # List all nodes
        elif "list nodes" in question:
            answer = list(G.nodes)

        # List all edges
        elif "list edges" in question:
            answer = list(G.edges)

        # Node neighbors
        elif "neighbors" in question:
            import re
            m = re.search(r"neighbors of ([^ ]+)", question)
            if m:
                node = m.group(1)
                if node in G.nodes:
                    answer = list(G.neighbors(node))
                else:
                    answer = f"Node {node} not found."
            else:
                answer = "Please specify the node (e.g., 'neighbors of AE_X')."

        # Default: not understood
        else:
            answer = "Sorry, I don't understand that question. Try 'highest weight', 'top 5 degree', 'list communities', etc."

        st.subheader("Answer")
        st.write(answer)
else:
    st.info("Please upload both nodes and edges JSON files.")
```
